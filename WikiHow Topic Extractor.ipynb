```python
import pandas as pd
import re
import pickle
import copy
from nltk.tokenize import sent_tokenize
```

```python
no_documents = 10000
```

```python
df = pd.read_csv("E:\\Bhargav\\wikihowAll.csv")
df.head(5)
```

```python
df_limited = df.head(no_documents)
```

```python
doc_id = []
doc_id_summary =[]
topic_id = []
topic_id_summary = []
summary = []
sentences_list = []
sentences_lists = []
sentence_id = []
no_fulltext_sentences = {}
no_summ_sentence = []
skipped_documents = []
doc_id_tracker = 0
```

```python
for itr in range(0, no_documents):
    # Structure of a dataframe: title, Headline: Summary, Text: Full length text
    # Checking whether full text or summary of topic is given or not.
    if (type(df_limited['text'][itr]) is not str) or (type(df_limited['headline'][itr]) is not str or (type(df_limited['title'][itr]) is not str)):
        skipped_documents.append(itr)
        continue
    if len(df_limited['text'][itr]) < 10 or len(df_limited['headline'][itr]) < 1:
        skipped_documents.append(itr)
        continue
    
    topic_title = df_limited["title"][itr]
    processed_text = df_limited['text'][itr].replace(".,",".")
    sentences = sent_tokenize(processed_text)
    
    pop_index = []
    
    for i in range(0,len(sentences)):
        if bool(re.findall('\w', sentences[i])) and len(sentences[i]) > 2:
            sentences[i] = sentences[i].replace("\n"," ")
            sentences[i] = sentences[i].replace("\'","'")
        else:
            print(doc_id_tracker)
            print(sentences[i])
            print(i)
            print("=======================================================")
            pop_index.append(i)
    
#     for sentence in sentences:
#         i = sentences.index(sentence)
        
#         if bool(re.findall('\w', sentence)) or len(sentence) > 2:
#             if doc_id_tracker == 196:
#                 print(sentence)
#                 print(i)
#                 print("=======================================================")
#             sentences[i] = sentences[i].replace("\n"," ")
#             sentences[i] = sentences[i].replace("\'","'")
#         else:
            
#             pop_index.append(i)
            
    pop_index.sort(reverse=True) 
    
    for i in pop_index:
        sentences.pop(i)
    
    if len(sentences) < 2:
        skipped_documents.append(itr)
        continue
        
    topic_id = topic_id + [topic_title]*len(sentences)
    doc_id += [doc_id_tracker]*len(sentences)
    sentences_list += sentences
    sentences_lists.append(sentences)
    no_fulltext_sentences[doc_id_tracker] = len(sentences)
    
    # Sentence ID generation
    temp = []
    for i in range(0,len(sentences)):
        temp.append('s' + str(i))
    sentence_id += temp
    
    # Summary dataframe
    topic_id_summary.append(topic_title)
    doc_id_summary.append(doc_id_tracker)
    processed_summary = str(df_limited['headline'][itr]).replace("\n"," ")
    processed_summary = processed_summary.replace(".,",". ")
    summary.append(processed_summary)
    no_summ_sentence.append(len(sent_tokenize(processed_summary)))
    doc_id_tracker += 1
```

```python
# Generates sentencewise CSV for each topic
dic = {'doc_id':doc_id, 'topic_id': topic_id, 'sen_id': sentence_id, 'sentence': sentences_list}
data_frame = pd.DataFrame(dic)
data_frame.reset_index(drop=True, inplace=True)
data_frame.to_csv("C:\\Users\\shirb\\Untitled Folder\\Major\\new\\dataset\\wikihow\\wiki_how_all_sentences.csv",index = False)
```

```python
# Generates summary topic mapping CSV
dic = {}
dic = {'doc_id':doc_id_summary, 'topic_id': topic_id_summary, 'summary': summary, 'sentence_count':no_summ_sentence}
data_frame = pd.DataFrame(dic)
data_frame.reset_index(drop=True, inplace=True)
data_frame.to_csv('C:\\Users\\shirb\\Untitled Folder\\Major\\new\\dataset\\wikihow\\wiki_how_all_summary.csv',index = False)
```
